## 📚 1. Importación de Librerías

```python
# Librerías para manipulación de datos
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

# Librerías para visualización
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Configuración de visualización
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

# Configuración de pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.float_format', '{:.2f}'.format)

# Suprimir warnings
import warnings
warnings.filterwarnings('ignore')

print("✅ Librerías importadas correctamente")
```

## 📂 2. Carga de Datos

```python
# Cargar el dataset de siniestros
df = pd.read_csv("Siniestros_nacionales_aleatorio_trim6.csv", encoding='latin-1', sep=';')

print(f"✅ Dataset cargado correctamente")
print(f"📊 Dimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")
```

```python
# Vista previa de los datos
print("\n📋 PRIMERAS 5 FILAS:")
df.head()
```

```python
# Información general del dataset
print("\n📊 INFORMACIÓN GENERAL:")
df.info()
```

## 🧹 3. Limpieza y Transformación de Datos

```python
# Crear una copia del dataframe original
df_original = df.copy()

# Limpiar nombres de columnas (eliminar espacios extras)
df.columns = df.columns.str.strip()

print("✅ Nombres de columnas limpiados")
```

```python
# Verificar y convertir tipos de datos
print("\n🔧 CONVERSIÓN DE TIPOS DE DATOS:")

# Convertir fecha a datetime si no lo está
if df['Fecha Siniestro'].dtype != 'datetime64[ns]':
    df['Fecha Siniestro'] = pd.to_datetime(df['Fecha Siniestro'])
    print("✅ Columna 'Fecha Siniestro' convertida a datetime")

# Verificar duplicados
duplicados = df.duplicated().sum()
print(f"\n🔍 Registros duplicados encontrados: {duplicados}")

if duplicados > 0:
    df.drop_duplicates(inplace=True)
    print(f"✅ {duplicados} duplicados eliminados")
```

```python
# Análisis de valores nulos
print("\n❌ VALORES NULOS POR COLUMNA:")
nulos = df.isnull().sum()
nulos_pct = (nulos / len(df)) * 100

nulos_df = pd.DataFrame({
    'Columna': nulos.index,
    'Valores Nulos': nulos.values,
    'Porcentaje': nulos_pct.values
}).sort_values('Valores Nulos', ascending=False)

nulos_df[nulos_df['Valores Nulos'] > 0]
```

```python
# Crear columnas derivadas para análisis temporal
print("\n🕐 CREANDO COLUMNAS TEMPORALES:")

df['Año'] = df['Fecha Siniestro'].dt.year
df['Mes'] = df['Fecha Siniestro'].dt.month
df['Mes_Nombre'] = df['Fecha Siniestro'].dt.month_name()
df['Día'] = df['Fecha Siniestro'].dt.day
df['Día_Semana'] = df['Fecha Siniestro'].dt.day_name()
df['Día_Semana_Num'] = df['Fecha Siniestro'].dt.dayofweek
df['Trimestre'] = df['Fecha Siniestro'].dt.quarter
df['Semana_Año'] = df['Fecha Siniestro'].dt.isocalendar().week
df['Semestre'] = df['Mes'].apply(lambda x: 'S1' if x <= 6 else 'S2')
df['Año_Mes'] = df['Fecha Siniestro'].dt.to_period('M')

print("✅ Columnas temporales creadas:")
print("   - Año, Mes, Mes_Nombre, Día")
print("   - Día_Semana, Día_Semana_Num")
print("   - Trimestre, Semestre, Semana_Año")
print("   - Año_Mes")
```

```python
# Crear categorías de gravedad basadas en tiempo inoperativo
print("\n⚠️ CREANDO CATEGORÍA DE GRAVEDAD:")

df['Gravedad'] = pd.cut(
    df['Tiempo Inoperativo ( HORAS )'],
    bins=[-np.inf, 0, 2, 24, 100, np.inf],
    labels=['Sin tiempo', 'Leve', 'Moderado', 'Grave', 'Crítico']
)

print("✅ Categoría de gravedad creada:")
print("   - Sin tiempo: 0 horas")
print("   - Leve: 1-2 horas")
print("   - Moderado: 3-24 horas")
print("   - Grave: 25-100 horas")
print("   - Crítico: >100 horas")

print("\n📊 Distribución de Gravedad:")
print(df['Gravedad'].value_counts())
```

```python
# Crear indicador de fin de semana
df['Es_FinDeSemana'] = df['Día_Semana_Num'].apply(lambda x: 'Sí' if x >= 5 else 'No')

# Crear indicador de horario laboral (aproximado)
# Asumimos que el turno nocturno es de mayor riesgo
df['Turno_Riesgo'] = df['Turno'].apply(lambda x: 'Alto' if x == 'Nocturno' else 'Normal')

print("\n✅ Variables adicionales creadas:")
print("   - Es_FinDeSemana")
print("   - Turno_Riesgo")
```

```python
# Resumen final del dataset limpio
print("\n" + "="*80)
print("📊 RESUMEN FINAL DEL DATASET LIMPIO")
print("="*80)
print(f"\n📏 Dimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")
print(f"📅 Periodo: {df['Fecha Siniestro'].min().strftime('%d/%m/%Y')} - {df['Fecha Siniestro'].max().strftime('%d/%m/%Y')}")
print(f"🗺️ Departamentos únicos: {df['DEPARTAMENTO'].nunique()}")
print(f"🏙️ Ciudades únicas: {df['CIUDAD'].nunique()}")
print(f"🚗 Tipos de vehículos: {df['Tipo de vehículo'].nunique()}")
print(f"👤 Conductores únicos: {df['Nombre Conductor'].nunique()}")
print(f"\n✅ Datos listos para análisis")
```

## 📊 4. Análisis Exploratorio de Datos (EDA)

### 4.1 Estadísticas Descriptivas Generales

```python
# Estadísticas de tiempo inoperativo
print("⏱️ ESTADÍSTICAS DE TIEMPO INOPERATIVO (HORAS):")
print("="*60)

tiempo_stats = df['Tiempo Inoperativo ( HORAS )'].describe()
print(tiempo_stats)

print(f"\n📊 Estadísticas adicionales:")
print(f"   Suma total: {df['Tiempo Inoperativo ( HORAS )'].sum():,.0f} horas")
print(f"   Rango: {df['Tiempo Inoperativo ( HORAS )'].max() - df['Tiempo Inoperativo ( HORAS )'].min():,.0f} horas")
print(f"   Percentil 90: {df['Tiempo Inoperativo ( HORAS )'].quantile(0.90):.2f} horas")
print(f"   Percentil 95: {df['Tiempo Inoperativo ( HORAS )'].quantile(0.95):.2f} horas")
```

```python
# KPIs principales
print("\n🎯 INDICADORES CLAVE DE DESEMPEÑO (KPIs):")
print("="*60)

total_siniestros = len(df)
conductor_culpable = (df['Culpabilidad Conductor'] == 'SI').sum()
pct_culpable = (conductor_culpable / total_siniestros) * 100
reportados = (df['Reportado Aseguradora'] == 'SI').sum()
pct_reportados = (reportados / total_siniestros) * 100
uso_grua = (df['Usó Grua'] == 'SI').sum()
pct_grua = (uso_grua / total_siniestros) * 100
asistencia_juridica = (df['Asistencia Jurídica'] == 'SI').sum()
pct_juridica = (asistencia_juridica / total_siniestros) * 100

print(f"📌 Total de Siniestros: {total_siniestros:,}")
print(f"👤 Conductor Culpable: {conductor_culpable:,} ({pct_culpable:.1f}%)")
print(f"📋 Reportados a Aseguradora: {reportados:,} ({pct_reportados:.1f}%)")
print(f"🚧 Uso de Grúa: {uso_grua:,} ({pct_grua:.1f}%)")
print(f"⚖️ Asistencia Jurídica: {asistencia_juridica:,} ({pct_juridica:.1f}%)")
print(f"⏱️ Tiempo Inoperativo Promedio: {df['Tiempo Inoperativo ( HORAS )'].mean():.2f} horas")
```

## 🚗 8. Análisis de Vehículos

```python
# Análisis por tipo de vehículo
print("\n🚗 ANÁLISIS POR TIPO DE VEHÍCULO:")
print("="*80)

vehiculo_stats = df.groupby('Tipo de vehículo').agg({
    'CIUDAD': 'count',
    'Tiempo Inoperativo ( HORAS )': ['mean', 'median'],
    'Culpabilidad Conductor': lambda x: (x == 'SI').sum() / len(x) * 100,
    'Usó Grua': lambda x: (x == 'SI').sum() / len(x) * 100,
    'Reportado Aseguradora': lambda x: (x == 'SI').sum() / len(x) * 100
}).round(2)

vehiculo_stats.columns = ['Total', 'Tiempo_Prom', 'Tiempo_Mediana', 
                          'Pct_Culpabilidad', 'Pct_Grua', 'Pct_Reportado']
vehiculo_stats = vehiculo_stats.sort_values('Total', ascending=False)
print(vehiculo_stats)
```

## 📊 11. Análisis Multivariado y Correlaciones

```python
# Análisis de tablas cruzadas
print("\n📊 TABLAS CRUZADAS:")
print("="*80)

print("\n1️⃣ Turno vs Culpabilidad:")
tabla1 = pd.crosstab(df['Turno'], df['Culpabilidad Conductor'], margins=True)
print(tabla1)

print("\n2️⃣ Departamento vs Causa (Top causas):")
tabla2 = pd.crosstab(df['DEPARTAMENTO'], df['Causa Siniestro'], margins=True)
print(tabla2)

print("\n3️⃣ Tipo de Vehículo vs Uso de Grúa (Top vehículos):")
top_vehiculos = df['Tipo de vehículo'].value_counts().head(5).index
tabla3 = pd.crosstab(
    df[df['Tipo de vehículo'].isin(top_vehiculos)]['Tipo de vehículo'],
    df[df['Tipo de vehículo'].isin(top_vehiculos)]['Usó Grua']
)
print(tabla3)
```

```python
# Heatmap de correlaciones entre variables categóricas
# Crear variables numéricas para correlación
df_corr = df.copy()
df_corr['Culpable_Num'] = (df_corr['Culpabilidad Conductor'] == 'SI').astype(int)
df_corr['Reportado_Num'] = (df_corr['Reportado Aseguradora'] == 'SI').astype(int)
df_corr['Grua_Num'] = (df_corr['Usó Grua'] == 'SI').astype(int)
df_corr['Juridica_Num'] = (df_corr['Asistencia Jurídica'] == 'SI').astype(int)
df_corr['Turno_Num'] = (df_corr['Turno'] == 'Nocturno').astype(int)

corr_cols = ['Tiempo Inoperativo ( HORAS )', 'Culpable_Num', 'Reportado_Num', 
             'Grua_Num', 'Juridica_Num', 'Turno_Num']

corr_matrix = df_corr[corr_cols].corr()

fig = go.Figure(data=go.Heatmap(
    z=corr_matrix.values,
    x=['Tiempo Inop', 'Culpable', 'Reportado', 'Grúa', 'Jurídica', 'Turno Noct'],
    y=['Tiempo Inop', 'Culpable', 'Reportado', 'Grúa', 'Jurídica', 'Turno Noct'],
    colorscale='RdBu',
    zmid=0,
    text=corr_matrix.values.round(2),
    texttemplate='%{text}',
    textfont={"size": 12}
))

fig.update_layout(
    title='Matriz de Correlaciones',
    height=600
)
fig.show()
```

## 📝 12. Conclusiones y Recomendaciones

### 🎯 Principales Hallazgos

**1. Distribución Geográfica:**
- Antioquia concentra el mayor número de siniestros (47.3%)
- Las ciudades de Ibagué, Bello y Valledupar lideran en incidentes

**2. Causas de Siniestros:**
- Los factores externos son la causa principal (54.5%)
- Las fallas humanas representan el 41.1% de los casos
- Las fallas de mantenimiento son minoritarias (4.4%)

**3. Patrones Temporales:**
- Mayor incidencia en turno diurno (60.7%)
- Los días laborables concentran más siniestros
- Se observa tendencia creciente en años recientes

**4. Vehículos:**
- Los compactadores están involucrados en el 71% de los casos
- Las volquetas tienen mayor tiempo inoperativo promedio

**5. Conductores:**
- El 44.9% de los casos tienen culpabilidad del conductor
- Existe un grupo de conductores reincidentes
